## Nested 중첩된 조건문

```python
year = 2025
month = 12
day = 31
day = day + 1
if day > 31 :
  month = month + 1
  day = 1

  if month > 12:
    year = year +1
    month = 1

print(f"{year}년, {month}월, {day}일")
```
