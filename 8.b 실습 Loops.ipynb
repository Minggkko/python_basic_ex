# <font color='blue'> 4. 1부터 100까지 출력하되, 3의 배수에는 "박수" 라는 글자를 출력하는 프로그램을 작성하세요.

```python
##### 아래처럼 출력됨. #####
for num in range(1,100+1):
  #3의 배수인가 판단, 참이면 => "박수" 출력
  if num %3 == 0 : #3으로 나눠서 떨어짐
    print("박수")
    else:
      print(num)
```

```python
#1부터 100까지 출력
list(range(1,100+1))
# 3의 배수에만 "박수"라는 글자 출력
for data in list(range(1,100+1)):
  if data % 3 == 0:
    print("박수")
  else:
    print(data)
```

# <font color='blue'> 9. 유저가 점수를 입력하면, 해당 학점을 출력하는 프로그램입니다.  
# <font color='blue'> 단, 유저가 -1 을 입력하면, 프로그램을 종료합니다. -1 을 입력하기 전에는 계속 입력을 받습니다.


- 사용자가 어떤 값을 입력함, 그 값에 따라 A, B, C, F로 등급을 표시함
- 숫자 <= 73 ; F
- 74 <= 숫자 < 84 : C
- 74 <= 숫자 < 94 : B
- 74 <= 숫자 < 100 : A

```python
# -1을 입력하기 전에는 계속 입력을 받습니다.
while True:
  score = int(input("점수 입력:"))
  #문자를 입력 했다면 ; pass
  if score == -1:
    break
  else:
    if score <= 73:
      print("F")
    elif score < 84:
      print("C")
    elif score < 94:
      print("B")
    else:
      print("A")
```

# <font color='blue'> 10. 유저한테 구구단 시작단을 입력받으세요.
# <font color='blue'> 2를 입력받으면, 2단에서 9단까지 출력하세요.
# <font color='blue'> 8을 입력받으면, 8단에서 9단까지 출력하세요.
# <font color='blue'> 음수 입력하거나 9단보다 큰 단을 입력하면, “잘못입력하였습니다.” 하고, 다시 단을 입력 받습니다.
# <font color='blue'> 단, 0 을 입력받으면, 프로그램을 종료하세요.
# <font color='blue'> 입력)  단을 입력하세요.
# <font color='blue'> 출력)  4X1 = 4
# <font color='blue'>          ...


```python
while True:
  num = int(input("단을 입력하세요:"))
  if num == 0:
    print("프로그램을 종료합니다.")
    break
  elif num < 2 or num > 9:
    print("잘못입력하였습니다.")
  else:
    for i in range(1, 10):
      print(f"{num} x {i} = {num*i}")
```
